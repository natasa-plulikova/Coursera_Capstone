# Load dataset

In this section we will load a dataset from wikipedia about neighborhoods in Toronto.

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wiki_df = pd.read_html(url)[0]
wiki_df
```

```python
# filter the dataframe for only values that have assigned Borough
toronto_df = wiki_df.loc[wiki_df['Borough'] != 'Not assigned'].reset_index(drop=True)
toronto_df.head(15)
```

```python
# check if the postal code values are unique
toronto_df['Postal Code'].is_unique
# since all postal codes are unique, no need for a groupping by Postal Code is needed
```

```python
#check Not assigned neighborhoods
toronto_df.loc[toronto_df['Neighbourhood'] == 'Not assigned']
# there are no not assigned neighbourhoods that will have a value in Borough.
# not assigned neighbourhoods were removed in the step that cleaned the data from not assigned Boroughs
```
