# Part 2: Attach geo data to the dataframe

```python
#define a function which returns the longitude and latitude
def findLongLat(postal_code):
    lat_lng_coords = None
    counter = 10 # we will do at most 10 attempts to avoid infinite loop
    while(lat_lng_coords is None and counter > 0):
        g = geocoder.google('{}, Toronto, Ontario'.format(postal_code))
        lat_lng_coords = g.latlng
        counter = counter - 1
    return lat_lng_coords
```

```python
#try geocoder to get the longitude and latitude with geocoder
toronto_df['LatLong'] = toronto_df.apply(lambda x: findLongLat(x['Postal Code']),axis = 1)
toronto_df.head()
```

```python
# the postal codes were not properly loaded, so proceed with the provided csv
import urllib.request
url = 'https://cocl.us/Geospatial_data'
filename = 'Geospatial_Coordinates.csv'
urllib.request.urlretrieve(url, filename)
geo_data = pd.read_csv('Geospatial_Coordinates.csv')
geo_data.head()
```

```python
# merge the two data frames
toronto_geo_data = toronto_df.merge(geo_data, on = 'Postal Code')
toronto_geo_data.head()
```

```python
# we drop the LatLong column that was not succesful from geocoder
toronto_geo_data.drop(columns = ['LatLong'], inplace = True)
toronto_geo_data.head()
```

```python
# !conda install -c conda-forge geopy --yes # uncomment if not installed
# !conda install -c conda-forge folium=0.5.0 --yes # uncomment if not installed
import config

import requests
from geopy.geocoders import Nominatim
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
import folium
```

```python
CLIENT_ID = config.CLIENT_ID
CLIENT_SECRET = config.CLIENT_SECRET
ACCESS_TOKEN = config.ACCESS_TOKEN
VERSION = '20180605' # Foursquare API version
LIMIT = 100 # A default Foursquare API limit value
radius = 500
print(CLIENT_ID)
print(CLIENT_SECRET)
print(ACCESS_TOKEN)
```

First, we get the nearby venues that are in each neighbourhood

```python
def getNearbyVenues(postal_codes, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for pc, lat, lng in zip(postal_codes, latitudes, longitudes):
        print(pc)
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&oauth_token={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            ACCESS_TOKEN,
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        urlrequest = requests.get(url).json()
        results = urlrequest["response"]['groups'][0]['items']

        # return only relevant information for each nearby venue
        venues_list.append([(
            pc, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Postal Code', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# sometimes there is a problem with foursquare API, so this step might need to be run multiple times
toronto_venues = getNearbyVenues(postal_codes = toronto_geo_data['Postal Code'], 
                                 latitudes = toronto_geo_data['Latitude'], 
                                 longitudes = toronto_geo_data['Longitude'])
toronto_venues.head()
```

```python
toronto_venues.shape
```

```python
#let's see which neighborhoods have highest number of venues
toronto_venues.groupby(['Postal Code']).count().sort_values(by = 'Venue', ascending = False).head(15)
```

```python
# create a one hot encoding of venue categories per each neighborhood
venue_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")
venue_onehot['Postal Code'] = toronto_venues['Postal Code']
```

```python
venue_groupped = venue_onehot.groupby('Postal Code').mean().reset_index()
venue_groupped.head()
```

```python
venue_groupped.shape
```

```python
# set index for postal code
venue_groupped.set_index('Postal Code', inplace = True)
venue_groupped.head()
```

```python
#perform kmeans clustering
from sklearn.cluster import KMeans
```

```python
# first determine how many clusters make sense through elbow method
Nclusters = 10
sum_squared_dist = []
for k in range(1,Nclusters):
    km = KMeans(n_clusters=k, random_state=0).fit(venue_groupped)
    sum_squared_dist.append(km.inertia_)
```

```python
# plot the clusters
plt.plot(range(1,Nclusters), sum_squared_dist, 'bx-')
plt.xlabel('number of clusters')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal number of clusters')
plt.show()
```

```python
# from the chart above, it seems 6 clusters is reasonable
kclusters = 6
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(venue_groupped)
```

```python
kmeans.labels_
```

```python
# add cluster label to the dataset
venue_groupped['Cluster Label'] = kmeans.labels_
venue_groupped.head()
```

```python
cluster_df = venue_groupped[['Neighborhood', 'Cluster Label']].reset_index(drop = True)
cluster_df.head()
cluster_df.shape
```

```python
toronto_merged = toronto_geo_data.set_index('Postal Code').merge(
    venue_groupped['Cluster Label'],
    left_index = True, 
    right_index = True)
toronto_merged.head()
```

Let's visualize the clusters on the map

```python
# first we get address of Toronto
address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('Latitude: %f, Longitude: %f' % (latitude, longitude))
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighbourhood'], toronto_merged['Cluster Label']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

Let's examine the clusters by looking at the most common venue category

```python
# function to return the most common values
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[0:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']


# create columns according to number of top venues
columns = ['Postal Code', 'Cluster Label']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Postal Code'] = venue_groupped.index.get_level_values('Postal Code')
neighborhoods_venues_sorted['Cluster Label'] = venue_groupped['Cluster Label'].values
venue_groupped['Cluster Label'].values

# we reset index in the venue_groupped dataset that was Postal Code and save in the database
venue_groupped_reset = venue_groupped.reset_index(drop = True)
for ind in venue_groupped_reset.index:
    neighborhoods_venues_sorted.iloc[ind, 2:] = return_most_common_venues(venue_groupped_reset.iloc[ind, :-1], num_top_venues)

neighborhoods_venues_sorted.head()
```
